# Quick Sort and Bubble Sort Overview

## Introduction

This document outlines the complexities associated with two common sorting algorithms: Quick Sort and Bubble Sort. It covers their time and space complexities in various scenarios.

## Complexity Definitions

- **Time Complexity**: Indicates the computational complexity that describes the amount of time an algorithm takes to complete as a function of the length of the input.
- **Space Complexity**: Represents the total amount of memory space required by the algorithm, including both the space taken by the inputs and any additional space used during the computation process.
- **Best Case**: The scenario under which the algorithm performs its function in the shortest amount of time.
- **Average Case**: The expected scenario under normal conditions, indicating the average performance of the algorithm.
- **Worst Case**: The scenario under which the algorithm performs its function in the longest amount of time.

## Algorithm Complexity Table

The following table provides a summary of the time and space complexities for Quick Sort and Bubble Sort algorithms:

| Algorithm   | Best Case Time Complexity | Average Case Time Complexity | Worst Case Time Complexity | Space Complexity     |
| ----------- | ------------------------- | ---------------------------- | -------------------------- | -------------------- |
| Quick Sort  | `O(n log n)`              | `O(n log n)`                 | `O(n^2)`                   | `O(log n)` to `O(n)` |
| Bubble Sort | `O(n)`                    | `O(n^2)`                     | `O(n^2)`                   | `O(1)`               |

## Conclusion

- **Quick Sort**:

  - Quick Sort is generally more efficient for large datasets due to its `O(n log n)` average time complexity.
  - The space complexity of Quick Sort can vary from `O(log n)` to `O(n)`, depending on the implementation and the depth of recursion.

- **Bubble Sort**:
  - Bubble Sort is a straightforward sorting algorithm that, despite its simplicity, becomes less efficient for larger datasets with its `O(n^2)` time complexity.
  - It is, however, very space-efficient, requiring only `O(1)` additional space.


```python
import timeit
import matplotlib.pyplot as plt
import random
import numpy as np

from data_structures_and_algorithms.sort import bubble_sort, quick_sort

default_repetitions = 50


def wrapper(func, *args, **kwargs):
    def wrapped():
        return func(*args, **kwargs)

    return wrapped


def timer(func, *args, repetitions):
    wrapped = wrapper(func, *args)
    times = timeit.repeat(wrapped, repeat=repetitions, number=1)
    return min(times) * 1000


def generate_random_list(n):
    return [random.randint(0, 1000) for _ in range(n)]


def generate_sorted_list(n):
    return [i for i in range(n)]


def generate_reversed_list(n):
    return [i for i in range(n, 0, -1)]
```

```python
def plot_average_time():
    n = np.linspace(50, 1000, 50, dtype=int)
    bubble_sort_times = []
    quick_sort_times = []

    for i in n:
        bubble_sort_time = timer(
            bubble_sort, generate_random_list(i), repetitions=default_repetitions
        )
        quick_sort_time = timer(
            quick_sort, generate_random_list(i), repetitions=default_repetitions
        )

        bubble_sort_times.append(bubble_sort_time)
        quick_sort_times.append(quick_sort_time)

    plt.plot(n, bubble_sort_times, label="Bubble Sort is O(n^2)")
    plt.plot(n, quick_sort_times, label="Quick Sort is O(n log(n))")
    plt.xlabel("n")
    plt.ylabel("Time (ms)")
    plt.legend()
    plt.title("Average Time of Bubble Sort and Quick Sort for Random List")
    plt.show()


plot_average_time()
```

```python
def plot_best_case():
    n = np.linspace(50, 1000, 50, dtype=int)
    bubble_sort_times = []
    quick_sort_times = []

    for i in n:
        bubble_sort_time = timer(
            bubble_sort, generate_sorted_list(i), repetitions=default_repetitions
        )
        quick_sort_time = timer(
            quick_sort, generate_sorted_list(i), repetitions=default_repetitions
        )

        bubble_sort_times.append(bubble_sort_time)
        quick_sort_times.append(quick_sort_time)

    plt.plot(n, bubble_sort_times, label="Bubble Sort is O(n)")
    plt.plot(n, quick_sort_times, label="Quick Sort is O(n log(n))")
    plt.xlabel("n")
    plt.ylabel("Time (ms)")
    plt.legend()
    plt.title("Best Case Time of Bubble Sort and Quick Sort for Sorted List")
    plt.show()


plot_best_case()
```

```python
def plot_worst_case():
    n = np.linspace(50, 1000, 50, dtype=int)
    bubble_sort_times = []
    quick_sort_times = []

    for i in n:
        bubble_sort_time = timer(
            bubble_sort, generate_reversed_list(i), repetitions=default_repetitions
        )
        quick_sort_time = timer(
            quick_sort, generate_reversed_list(i), repetitions=default_repetitions
        )

        bubble_sort_times.append(bubble_sort_time)
        quick_sort_times.append(quick_sort_time)

    plt.plot(n, bubble_sort_times, label="Bubble Sort is O(n^2)")
    plt.plot(n, quick_sort_times, label="Quick Sort is O(n log(n))")
    plt.xlabel("n")
    plt.ylabel("Time (ms)")
    plt.legend()
    plt.title("Worst Case Time of Bubble Sort and Quick Sort for Reversed List")
    plt.show()


plot_worst_case()
```
